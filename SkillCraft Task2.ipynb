```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (update path as needed)
df = pd.read_csv(r"C:\Users\aksha\OneDrive\Desktop\titanic\train.csv")

# Set theme
sns.set(style="whitegrid")

# Create bar plot: survival count by gender
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set2')

# Add labels
plt.title("Titanic Survival Count by Gender", fontsize=16, weight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Passenger Count", fontsize=12)
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Chart 2: Survival Count by Passenger Class
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set1')

plt.title("Survival Count by Passenger Class", fontsize=14, weight='bold')
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Passenger Count", fontsize=12)
plt.legend(title="Survived", labels=["No", "Yes"])
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the cleaned CSV file
df = pd.read_csv("train_clean.csv")  # make sure this file exists in same folder as notebook

# Convert Age column to numeric in case of any stray non-numeric values like 'Q'
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Drop missing Age values
df_clean = df[df["Age"].notnull()]

# Plot KDE of Age by Survival
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df_clean,
    x="Age",
    hue="Survived",
    fill=True,
    alpha=0.6,
    linewidth=2
)

plt.title("Age Distribution by Survival", fontsize=14, weight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
